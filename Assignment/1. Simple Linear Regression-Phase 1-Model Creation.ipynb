```python
#pandas library used to read data from the csv file, we have imported pandas library as pd
import pandas as pd
```

```python
# we read the data inside the Salary_Dataset CSV file
dataset = pd.read_csv("Salary_Dataset.csv")
```

```python
dataset
```

```python
# we are getting the independent Columen data Years Of Experience and store it in the independent variable
independent = dataset[["Years Of Experience"]]
```

```python
independent
```

```python
# we are getting the dependent Columen data Salary and store it in the independent variable
dependent = dataset[["Salary"]]
```

```python
dependent
```

```python
# we are splitting the data into train and test sets using the sklearn.model_selection module, with the test set size as 30% and random state as 0
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
# we are importing the LinearRegression class from the sklearn library and we are creating the regressor which will be creating the model using the train
#sets, fit function is used to create the model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#slope value is calculated
weight = regressor.coef_
weight
```

```python
#bias value is calculated
bias=regressor.intercept_
bias
```

```python
#prediction values are calculated using the test input set
y_predict = regressor.predict(x_test)
```

```python
#using the r2 validation as the evaluation metrice, the prediction examination is done the value will lie between 0 and 1. 
#if the value is near to 1 the model has better performance, else the model has poor performance
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_predict)
```

```python
r_score
```

```python
#below is for the deployment phase, we will be dumping the regressor model in a file using this library.
import pickle
filename = "finalized_modal_linear.sav"
```

```python
# the model is dumbed into the finalized_modal_linear file
pickle.dump(regressor, open(filename, 'wb'))
```

```python
#when ever to check for any input with the model, we need to read the file and pass the input to predict method from the modal
loaded_modal= pickle.load(open(filename, 'rb'))
result=loaded_modal.predict([[14]])
```
