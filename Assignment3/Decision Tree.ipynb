```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset_encoded = pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
independent = dataset_encoded[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent = dataset_encoded[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(criterion="poisson" , splitter="random")
```

```python
model = regressor.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(model)
plt.show()
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```
