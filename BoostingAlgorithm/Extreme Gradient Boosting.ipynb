```python
import xgboost as xgb
```

```python
import pandas as pd
```

```python
dataset = pd.read_csv("Startups.csv")
```

```python
dataset_encoded = pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
independent = dataset_encoded[['R&D Spent', 'Administration Spent', 'Marketing Spent', 'State_Dehli', 'State_Kolkata', 'State_Mumbai']]
```

```python
dependent = dataset_encoded[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
regressor = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```
